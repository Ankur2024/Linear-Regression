**Importing the libraries and datset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

salary = pd.read_csv("Salary_Data.csv")
x = np.array(salary[['YearsExperience']])
y = np.array(salary[['Salary']])

plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("X And Y Data")
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print("Mean for x values is",x_mean)
print("Mean for y values is",y_mean)
```

**To Calculate value of coeffcient and intercept**

```python
num=0
den=0
for i in range(len(x)):
  num += (x[i]-x_mean)*(y[i]-y_mean)
  den += (x[i]-x_mean)**2

b1 = num/den
b0 = y_mean - (b1*x_mean)
print("The Intercept is ", float(b0[0]))
print("The coefficient is", float(b1[0]))
```

```python
y_preds=b0+b1*x
y_preds
```

```python
plt.scatter(x,y)
plt.plot(x,y_preds,color='red')
plt.show()
```

R-square = 1-(SSR/SST)

```python
num_r=0
den_r=0
for i in range(len(x)):
  num_r += (y[i]-y_preds[i]) ** 2
  den_r += (y[i]-y_mean) ** 2

r_square = 1-(num_r/den_r)
print("R-square is",float(r_square[0]))
```
